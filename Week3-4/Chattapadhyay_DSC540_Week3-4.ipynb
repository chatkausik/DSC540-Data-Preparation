```python
# 0.0.3 Load the necessary libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 0.0.4 Read in the Boston housing dataset (given as a .csv file) from the local directory.

bostonHousingDF=pd.read_csv("Boston_housing.csv")
```

```python
# 0.0.5 Check first 10 records

bostonHousingDF.head(10)
```

```python
# 0.0.6 Find the total number of records.

bostonHousingDF.shape
```

```python
# 0.0.7 Create a smaller DataFrame with columns which do not include ‘CHAS’, ‘NOX’, ‘B’, and ‘LSTAT’

# bostonHousingSmallDF = bostonHousingDF[['CRIM','ZN','INDUS','RM','AGE','DIS','RAD','TAX','PTRATIO','PRICE']]

bostonHousingSmallDF = bostonHousingDF.drop(["CHAS", "NOX", "B", "LSTAT"], axis=1)

# 0.0.8 Check the last 7 records of the new DataFrame you just created
bostonHousingSmallDF.tail(7)
```

```python
# 0.0.9 Can you plot histograms of all the variables (columns) in the new DataFrame?

for c in bostonHousingSmallDF.columns:
    plt.title("Plot of "+c,fontsize=15) 
    plt.grid(True)
    plt.xlabel(c,fontsize=15)
    plt.ylabel("Frequency",fontsize=15)
    plt.hist(bostonHousingSmallDF[c],bins=20) 
    plt.show()
```

```python
# 0.0.10 Crime rate could be an indicator of house price (people don’t want to live in high-crime areas). 
# Create a scatter plot of crime rate vs. Price.

plt.scatter(x=bostonHousingSmallDF['CRIM'],y=bostonHousingSmallDF['PRICE']) 
plt.title("Crime Rate Vs Housing Price Plot", fontsize=15)
plt.xlabel("Crime Rate", fontsize=15)
plt.ylabel("Housing Price", fontsize=15)
plt.grid(True)
plt.show()
```

```python
# 0.0.11 We can understand the relationship better if we plot log10(crime) vs. Price

plt.scatter(np.log10(bostonHousingSmallDF['CRIM']),bostonHousingSmallDF['PRICE'],c='red') 
plt.title("Crime rate (Log) vs. Price plot", fontsize=18) 
plt.xlabel("Log of Crime rate",fontsize=15) 
plt.ylabel("Price",fontsize=15) 
plt.grid(True) 
plt.show()
```

```python
# 0.0.12 Calculate the mean rooms per dwelling?

bostonHousingSmallDF['RM'].mean()
```

```python
# 0.0.13 Calculate median Age?

bostonHousingSmallDF['AGE'].median()
```

```python
# 0.0.14 Calculate average (mean) distances to five Boston employment centres?

bostonHousingSmallDF['DIS'].mean()
```

```python
#Load the necessary libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read in the adult income data set (given as a .csv file) from the local directory and check first 5 records

df = pd.read_csv("adult_income_data.csv", header=None) 
df.head()
```

```python
# Create a script that will read a text file line by line and extracts the first line, 
# which is the header of the .csv file:

names=[]
with open('adult_income_names.txt','r') as f:
    for line in f:
        f.readline()
        var1=line.split(":")[0]
        names.append(var1)
        
names          
```

```python
# Add a name (“Income”) for the response variable (last column) to the dataset and 
# read it again with the column names supplied.

names.append('Income')
names
```

```python
df.columns = names
```

```python
df.head()
```

```python
# Show a statistical summary of the data set. Did you notice only a small number of columns are included?

df.describe()
```

```python
df.info()
```

```python
# Find the missing values.

df.isna().sum()

# No missing values
```

```python
# Create a DataFrame with only age, education, and occupation by using subsetting:

df_subset = df[['age','education','sex', 'occupation']] 
df_subset.head()
```

```python
# Show the histogram of age with bin size = 20

df_subset['age'].hist(bins=20)
plt.xlabel("Age", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.title("Histogram of Age", fontsize=15)
plt.show()
```

```python
# Show boxplots of age grouped by race (Use a long figure size 15x6 and make x ticks font size 15 )

df_subset.boxplot(column='age',by='sex',figsize=(15,6)) 
plt.xticks(fontsize=15) 
plt.ylabel("Age",fontsize=20) 
plt.xlabel("Sex",fontsize=20)
plt.show()
```

```python
# Create a function to strip the whitespace characters

def strip_whitespace(s): 
    return s.strip()
```

```python
# Use the ‘apply’ method to apply this function to all the columns with string values, create a new column, 
# copy the values from this new column to the old column, and drop the new column.

for col in df.select_dtypes(include=object).columns:
    df[col+'_stripped'] = df[col].apply(strip_whitespace)
    df[col]=df[col+'_stripped'] 
    df.drop(labels=[col+'_stripped'],axis=1,inplace=True)

df.head()    
```

```python
# Find the number of people who are aged between 30 and 50 (inclusive) by using the following command

df_filtered = df[(df['age'] >= 30) & (df['age'] <= 50)]
df_age = pd.DataFrame(df_filtered['age'].value_counts().sort_index().reset_index())
df_age.columns = ['age', 'count']
print("There are {} people of age between 30 and 50 in this dataset.".format(df_age['count'].sum()))
```

```python
# Group the records based on age and education to find how the mean age is distributed:

byAgeEducation= df.groupby(['education'])
pd.DataFrame(byAgeEducation.describe()['age'])['mean'].sort_values()
```

```python
# Group the records based on occupation to find how the mean age is distributed:

pd.DataFrame(df.groupby('occupation').describe()['age'])['mean'].sort_values()
```

```python
# Group by occupation and show the summary statistics of age. Find which profession has the oldest workers on 
# average and which profession has its largest share of workforce above the 75th percentile:

pd.DataFrame(df.groupby('occupation').describe()['age'])['75%'].sort_values()

# Exec-managerial occupation has oldest workers and Priv-house-serv has it's largest share of workforce above the 
# 75th percentile.
```

```python
# Use subset and groupby to find the outliers

occupation_stats= df_subset.groupby( 'occupation').describe()['age']
occupation_stats
```

```python
# Plot the values on a bar chart

plt.figure(figsize=(15,8)) 
plt.barh(y=occupation_stats.index, width=occupation_stats['count']) 
plt.yticks(fontsize=13) 
plt.show()
```

```python
# Practice Merging by common keys: Suppose you are given two datasets where the common key is occupation. 
# Can you merge them?

df_1 = df[['age','workclass','occupation']].sample(5,random_state=101) 
df_1.head()
```

```python
df_2 = df[['education','sex','occupation']].sample(5,random_state=101) 
df_2.head()
```

```python
df_merged = pd.merge(df_1,df_2,on='occupation',how='inner').drop_duplicates() 
df_merged
```

# Create a series and practice basic arithmetic steps

1. Series 1 = 7.3, -2.5, 3.4, 1.5

2. Index = ‘a’, ‘c’, ‘d’, ‘e’

3. Series 2 = -2.1, 3.6, -1.5, 4, 3.1

4. Index = ‘a’, ‘c’, ‘e’, ‘f’, ‘g’

5. Add Series 1 and Series 2 together and print the results

6. Subtract Series 1 from Series 2 and print the results

```python
# series 1

data1 = np.array([7.3, -2.5, 3.4, 1.5])

# providing an index

series1 = pd.Series(data1, index =['a', 'c', 'd', 'e'])

print(series1)
```

```python
# series 2

data2 = np.array([-2.1, 3.6, -1.5, 4, 3.1])

# providing an index 

series2 = pd.Series(data2, index =['a', 'c', 'e', 'f','g']) 
print(series2)
```

```python
# series 1 + series 2 

series3 = series1.add(series2, fill_value=10); 
print(series3)
```

```python
# series 1 - series 2 

series4 = series1.subtract(series2, fill_value=10); 
print(series4)
```
