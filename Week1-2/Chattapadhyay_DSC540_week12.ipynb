**3.a. Data Wrangling with Python: Activity 1 page 17** 

1. Create a list of 100 random numbers. 
2. Create a new list from this random list, with numbers that are divisible by 3.
3. Calculate the length of these two lists and store the difference in a new variable.
4. Using a loops perform steps 2, 3 and find the difference variable three times.
5. Find the arithmetic mean of these three different values.

```python
def list_calculation(limit, divisibleBy):
    random_list = [random.randint(0,100) for x in range(0, limit)]
    divisible_by_3_list = [a for a in random_list if a % divisibleBy == 0]
    print("\n List Data \n")
    print(random_list)
    print("\n Divisible by 3 \n")
    print(divisible_by_3_list)
    random_list_len = len(random_list)
    random_list_div3_len = len(divisible_by_3_list)
    diff_len = random_list_len - random_list_div3_len
    return diff_len
```

```python
diffoflist = list_calculation(100, 3)
print("\n Difference of List \n")
print(diffoflist)

repeatCount = 3
diff_list = []
for i in range(0, repeatCount):
    diff_list.append(list_calculation(100, 3))

print("\n Difference of List \n")
print(diff_list)

avg_diff = sum(diff_list) / float(len(diff_list))
print("\n Average of difference \n")
print(avg_diff)
```

**3.d. Data Wrangling with Python: Activity 4 page 59** 
1. Import zip_longest from itertools. Create a function to zip header, line and fillvalue=None. 
2. Open the accompanying sales_record.csv file from the GitHub link by using r mode inside a with block and first check that it is opened. 
3. Read the first line and use string methods to generate a list of all the column names. 
4. Start reading the file. Read it line by line. 
5. Read each line and pass that line to a function, along with the list of the
headers. The work of the function is to construct a dict out of these two and fill up the key:values.
Keep in mind that a missing value should result in None.

```python
from itertools import zip_longest
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # looping over the first 10 lines in order to not to make the output too big
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
             break
```
